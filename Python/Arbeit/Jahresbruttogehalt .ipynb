```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Monatsbrutto nach IGM
IAVList = []
IAVList.append(['IAV start',4633]) 
IAVList.append(['IAV E20',5638]) 
IAVList.append(['IAV E21',5819]) 
IAVList.append(['IAV E22',5925]) 
IAVList.append(['IAV E23',6096]) 
IAVList.append(['IAV E24',6253])
IAVList.append(['IAV E25',6421]) 

# Monatsbrutto nach IGM
IGMList=[]
IGMList.append(['IAV start',4633]) 
IGMList.append(['EG10 A',4263]) 
IGMList.append(['EG11 A',4432]) 
# IGMList.append(['EG11 C',4851]) 
IGMList.append(['EG12 A',5056]) 
IGMList.append(['EG12 B',5262]) 
IGMList.append(['EG12 C',5471]) 
IGMList.append(['EG13 A',5704]) 
# IGMList.append(['EG13 B',6152]) 
IGMList.append(['EG13 C',6335]) 

#Jahresbrutto
fac_bonus = 76.0/67.0*12.0
fac_urlaub = 0.7
fac_bonus = 12.0 + fac_urlaub
fac_part_time = 30.0/40.0
fac_part_time = 30.0/38.0
print(f'fac_bonus = {fac_bonus}')
# fac_bonus = 67/59*12
# print(f'fac_bonus = {fac_bonus}')

BAV = 12.0*275.0




```

```python
# Define Brutto List
BruttoList=[]
BruttoList.append(['2016',41882]) 
BruttoList.append(['2017',48665]) 
BruttoList.append(['2018',46665]) 
BruttoList.append(['2019',55009]) 
BruttoList.append(['2020',51596]) 
BruttoList.append(['2021',50053]) 
BruttoList.append(['2022',54988]) 
```

```python
fac_bonus = 76.0/67.0*12.0
fac_urlaub = 0.7
fac_bonus = 12.0 + fac_urlaub

# create data 
BruttoIAV = []
for item in IAVList:
    fac_bonus = 12.0 
    BruttoIAV.append([item[0],item[1]*fac_bonus*fac_part_time])

# Create figure without data 
fig, axs = plt.subplots(1, 1)
fig.set_figwidth(12)   
fig.set_figheight(6)   
axs.grid(True)
axs.set_title(f'IAV Tarif 30H : fac_part_time={fac_part_time} / fac_bonus={fac_bonus}')
axs.set_xlabel('year')
axs.set_ylabel('brutto [€]')
axs.set_ylim([40000,90000])

# get x/y data
x = []
y_true=[]
for item in BruttoList:
    x.append(int(item[0]))
    y_true.append(item[1]+BAV)

# add plots
axs.plot(x,y_true,label='true')

# add IGM levels
x = []
x.append(int(BruttoList[0][0]))
x.append(int(BruttoList[-1][0]))

for item in BruttoIAV:
    axs.plot(x,[item[1],item[1]],label=item[0])

# show figure
axs.legend()
plt.show()
```

```python
# IGM 2023 Weihnachtsgeld & Urlaub --> 110%
fac_part_time = 1.0
fac_urlaub = 0.7
fac_bonus = 12.0 + fac_urlaub +0.4

# Create Data
BruttoIGM = []
for item in IGMList:
    BruttoIGM.append([item[0],item[1]*fac_bonus*fac_part_time])

# Create figure without data 
fig, axs = plt.subplots(1, 1)
fig.set_figwidth(12)   
fig.set_figheight(6)   
axs.grid(True)
axs.set_title(f'IGM Compare - 38 hours : fac_part_time={fac_part_time} / fac_bonus={fac_bonus}')
axs.set_xlabel('year')
axs.set_ylabel('brutto [€]')
axs.set_ylim([40000,90000])

# get x/y data
x = []
y_true=[]
for item in BruttoList:
    x.append(int(item[0]))
    y_true.append(item[1]+BAV)

# add plots
axs.plot(x,y_true,label='true')

# add IGM levels
x = []
x.append(int(BruttoList[0][0]))
x.append(int(BruttoList[-1][0]))

axs.plot(x,[5925*12,5925*12],label='IAV E22')

for item in BruttoIGM:
    axs.plot(x,[item[1],item[1]],label=item[0])

# show figure
axs.legend()
plt.show()
```
